```python
#ex 3
N=40
r=4
x=zero_vector(RR, N+1)
x[0] = 0.67

y=zero_vector(RR, N+1)
y[0]=0.670001

for n in range(1, N + 1):
    x[n]=r*x[n-1]*(1-x[n-1])
    y[n]=r*y[n-1]*(1-y[n-1])
x, y
```

```python
lx=line(Line(range(N), x, {}), color='blue', marker='o')
ly=line(Line(range(N), y, {}), color='red', marker='o')
show(lx+ly)
```

```python
#ex 4
N=11
h=1/(N-1)

x=zero_vector(RR, N)
y=zero_vector(RR, N)
x[0]=0
y[0]=1

ye=zero_vector(RR, N)
ye[0]=1

for k in range(1, N):
    x[k]=x[k-1]+h
    y[k]=y[k-1]+h*2*x[k-1]*y[k-1]
    ye[k]=exp(x[k]^2)
    
ly=line(Line(range(N), y, {}), color='red', marker='o')
lye=line(Line(range(N), ye, {}), color='blue', marker='o')
show(ly+lye)
```

```python
#y'=ay, y(0)=1
a=22
h=0.1
N=11
x=zero_vector(RR, N)
y=zero_vector(RR, N)
yy=zero_vector(RR, N)
y[0]=1
yy[0]=1

for k in range(1, N):
    x[k]=x[k-1] + h
    y[k]=y[k-1] + h*a*y[k-1]
    yy[k]=yy[k-1]+h/2*a*yy[k-1] + h/2*h*a*(yy[k-1] + h*a*yy[k-1])

ly=line(Line(x, y, {}), color='blue', marker='o')
lyy=line(Line(x, yy,{}), color='red', marker='o')
show(ly+lyy)
```
